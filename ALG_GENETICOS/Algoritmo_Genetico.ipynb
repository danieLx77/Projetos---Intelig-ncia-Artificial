## Introdução

Este notebook apresenta a construção dos componentes centrais de um **Algoritmo Genético (AG)** para encontrar o valor máximo da função:

$$f(x) = x^2 - 3x + 4$$

dentro do intervalo:

$$x \in [-10, 10]$$

A implementação utiliza uma representação binária para codificar os indivíduos da população. A partir dessa codificação, são definidos os parâmetros do algoritmo, a função de avaliação e os operadores genéticos necessários para evolução da população.

A estrutura segue os requisitos do enunciado:

- população inicial configurável, começando com 4 indivíduos;

## 1. Importações e Parâmetros do Algoritmo

Nesta etapa importamos as bibliotecas necessárias e definimos os principais parâmetros do Algoritmo Genético.

Os valores foram organizados como variáveis para facilitar os testes com diferentes configurações, conforme solicitado no enunciado da atividade.

Entre os parâmetros definidos estão:

- tamanho da população;
- número de gerações;
- taxa de crossover;
- taxa de mutação;
- quantidade de bits do cromossomo;
- intervalo permitido para o valor de $x$;
- função objetivo a ser maximizada.

```python
import random
import matplotlib.pyplot as plt

# 1. Parâmetros Variáveis 
TAMANHO_POPULACAO = 4
GERACOES = 5
TAXA_CROSSOVER = 0.70  # 70%
TAXA_MUTACAO = 0.01    # 1%

# Parâmetros da Codificação do Cromossomo
NUM_BITS = 10 
LIMITE_INF = -10
LIMITE_SUP = 10

FUNCAO = "x**2 - 3*x + 4"

print("Bibliotecas importadas e parâmetros configurados com sucesso!")
```

## 2. Codificação Binária, Decodificação e Fitness

Cada indivíduo da população é representado por um cromossomo binário. Esse cromossomo precisa ser convertido para um valor real dentro do intervalo $[-10, 10]$.

A função de decodificação transforma o vetor binário em um número inteiro e, em seguida, mapeia esse valor para o intervalo definido no problema.

Depois da decodificação, calculamos o **fitness** do indivíduo aplicando o valor de $x$ na função objetivo:

$$f(x) = x^2 - 3x + 4$$

Como o objetivo é maximizar a função, indivíduos com maior valor de fitness são considerados melhores soluções.

```python
# 2. Funções de Decodificação e Fitness

def decodificar(cromossomo):
    valor_inteiro = int("".join(str(b) for b in cromossomo), 2)
    valor_maximo_binario = (2 ** NUM_BITS) - 1
    
    x = LIMITE_INF + (valor_inteiro / valor_maximo_binario) * (LIMITE_SUP - LIMITE_INF)
    return x

def calcular_fitness(cromossomo):
    x = decodificar(cromossomo)
    resultado = eval(FUNCAO)
    return resultado

print("Motor de tradução genética e cálculo de aptidão inicializado!")
```

## 3. Operadores Genéticos

Nesta etapa implementamos os principais operadores do Algoritmo Genético:

- **População inicial:** cria cromossomos binários aleatórios;
- **Seleção por torneio:** escolhe os melhores indivíduos entre competidores aleatórios;
- **Crossover:** combina partes de dois pais para gerar novos filhos;
- **Mutação:** altera bits do cromossomo com baixa probabilidade.

Esses operadores permitem que a população seja modificada ao longo das gerações, explorando diferentes regiões do espaço de busca.

```python
# 3. Operadores Genéticos (Torneio, Crossover e Mutação)

def criar_populacao_inicial(tamanho):
    populacao = []
    for _ in range(tamanho):
        cromossomo = [random.randint(0, 1) for _ in range(NUM_BITS)]
        populacao.append(cromossomo)
    return populacao

def selecao_torneio(populacao):
    competidores = random.sample(populacao, 3)
    vencedor = max(competidores, key=calcular_fitness)
    return vencedor

def crossover(pai1, pai2):
    if random.random() < TAXA_CROSSOVER:
        ponto_corte = random.randint(1, NUM_BITS - 1)
        filho1 = pai1[:ponto_corte] + pai2[ponto_corte:]
        filho2 = pai2[:ponto_corte] + pai1[ponto_corte:]
        return filho1, filho2
    else:
        return pai1[:], pai2[:]

def mutacao(cromossomo):
    for i in range(len(cromossomo)):
        if random.random() < TAXA_MUTACAO:
            cromossomo[i] = 1 - cromossomo[i]
    return cromossomo

print("Seleção, Crossover e Mutação implementados! O DNA está pronto para evoluir.")
```
